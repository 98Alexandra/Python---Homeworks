```python
import numpy
import matplotlib.pyplot as plt

# Generowanie losowych punktów
n = 100
x = numpy.random.rand(n)
y = numpy.random.rand(n)

# Obliczanie odległości od punktu (0,0)
distances = numpy.sqrt(x**2 + y**2)

# Kolorowanie punktów na podstawie odległości
colors = numpy.where(distances < 1, 'green', 'red')

# Rozmiar znacznika punktu proporcjonalny do |x|+|y|
sizes = numpy.abs(x) + numpy.abs(y)

# Wykreślenie punktów
plt.scatter(x, y, c=colors, s=sizes*100)

# Ustawienie granic osi
plt.xlim(0, 1)
plt.ylim(0, 1)

# Dodanie tytułu i etykiet osi
plt.title('Random Points in Unit Square')
plt.xlabel('x')
plt.ylabel('y')

# Wyświetlenie wykresu
plt.show()
```
