```python
import random

def random_walk(start):
    position = start
    for _ in range(100):  # Powtarzamy ruchy 100 razy
        # Losujemy, czy poruszyć się w lewo czy w prawo
        step = random.choice([-1, 1])
        position += step
        yield position  # Zwracamy aktualną pozycję


start_position = 0
walker = random_walk(start_position)

for i, position in enumerate(walker):
    print(f"Position after {i+1} steps: {position}")
```
