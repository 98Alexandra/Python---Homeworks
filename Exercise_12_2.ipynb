```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Tworzenie losowego grafu nieskierowanego
def create_random_graph(num_nodes, num_edges):
    G = nx.Graph()
    G.add_nodes_from(range(num_nodes))

    while len(G.edges) < num_edges:
        u = random.choice(range(num_nodes)) #wierzchołek 1
        v = random.choice(range(num_nodes)) #wierzchołek 2
        if u != v and not G.has_edge(u, v): #dodaj krawędz od 1 do 2
            G.add_edge(u, v)

    return G

num_nodes = 6 # liczba punktów
num_edges = 8  # liczba krawędzi, musi być >= num_nodes - 1 dla spójnego grafu
G = create_random_graph(num_nodes, num_edges)

# Rysowanie grafu
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=700, edge_color='gray', font_size=15, font_color='black', font_weight='bold')
plt.title("Random Undirected Graph")
plt.show()
```
